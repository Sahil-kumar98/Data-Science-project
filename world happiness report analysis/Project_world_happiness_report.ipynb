### Top 10 Highest and Lowest Happiness Scores 2015 to 2020

### since we have different columns name for same varibles in reports for all different year. example -  Happiness Score,Ladder score,score are different name columns for same varibles 'Happiness Score'. to make thing easy , i am making same  name columns for a given variable .
 

```python
happy_2017.rename(columns={'Country or region':'Country', 'Happiness.Score':'Happiness Score'},inplace = True)
happy_2018.rename(columns={'Country or region':'Country', 'Score':'Happiness Score'},inplace = True)
happy_2019.rename(columns={'Country or region':'Country', 'Score':'Happiness Score'},inplace = True)
happy_2020.rename(columns={'Country name':'Country' , 'Ladder score':'Happiness Score' },inplace = True)
```

```python
df_lists = [happy_2015,happy_2016,happy_2017,happy_2018,happy_2019,happy_2020]
```

```python
for i,j in zip(range(2015,2021),df_lists) :
    fig = plt.figure(figsize=(12,4))
    ax1 = fig.add_subplot(1,2,1)
    ax2 = fig.add_subplot(1, 2, 2)
    
    ax1.barh(j['Country'].head(10), j['Happiness Score'].head(10),color ='g')
    ax1.set_title("Top 10 happiest countries in "+ str(i))
    for index, value in zip(range(10),j['Happiness Score'].head(10)):
         ax1.text(value, index, str(round(value,2)))

    
    ax2.barh(j['Country'].tail(10), j['Happiness Score'].tail(10))
    ax2.set_title("Least 10 happiest countries in " + str(i) )
    fig.tight_layout()
    for index, value in zip(range(10),j['Happiness Score'].tail(10)):
        ax2.text(value, index, str(round(value,2)))
    fig.tight_layout()
    
```

## Conclusion :

- we have got,Happiness score of a country is positivly related to GDP per capita, Healthy life expectancy, Freedom to make life choices but least associted or idependent with Generosity and Perception of corruption indicators.
